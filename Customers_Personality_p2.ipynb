## Clustering Analysis

```python
from matplotlib import colors
from matplotlib.colors import ListedColormap
import plotly as py
import plotly.graph_objs as go
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
from  sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
#Creating a copy of data
data = df.copy()
# creating a subset of dataframe by dropping the features on deals accepted and promotions
cols_del = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2', 'Complain', 'Response']
data = data.drop(cols_del, axis=1)
#Scaling
scaler = StandardScaler()
# Fit the scaler on your data, not 'ds' which seems to be a plot object
scaler.fit(data)
scaled_data = pd.DataFrame(scaler.transform(data),columns= data.columns )
print("All features are now scaled")
```

```python
#Initiating PCA to reduce dimentions aka features to 3
pca = PCA(n_components=3)
pca.fit(scaled_data)
PCA_data = pd.DataFrame(pca.transform(scaled_data), columns=(["col1","col2", "col3"]))
PCA_data.describe().T
```

```python
#A 3D Projection Of Data In The Reduced Dimension
x =PCA_data["col1"]
y =PCA_data["col2"]
z =PCA_data["col3"]
#To plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z, c="maroon", marker="o" )
ax.set_title("A 3D Projection Of Data In The Reduced Dimension")
plt.show()
```

```python
# Quick examination of elbow method to find numbers of clusters to make.
!pip install yellowbrick
#Import the necessary library
from yellowbrick.cluster import KElbowVisualizer

# Quick examination of elbow method to find numbers of clusters to make.
print('Elbow Method to determine the number of clusters to be formed:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(PCA_data)
Elbow_M.show()
```

```python
#Initiating the Agglomerative Clustering model
AC = AgglomerativeClustering(n_clusters=5)
# fit model and predict clusters
yhat_AC = AC.fit_predict(PCA_data)
PCA_data["Clusters"] = yhat_AC
#Adding the Clusters feature to the orignal dataframe.
data["Agg_Clusters"]= yhat_AC
```

```python
#Plotting the clusters
import matplotlib.pyplot as plt
from matplotlib import cm

fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(x, y, z, s=40, c=PCA_data["Clusters"], marker='o', cmap = cm.viridis ) # Use a colormap like 'viridis'
ax.set_title("The Plot Of The Clusters")
plt.show()
```

```python
#Plotting countplot of clusters
pal = ["#682F2F","#B9C0C9", "#9F8A78","#F3AB60"]
pl = sns.countplot(x=data["Agg_Clusters"], palette= pal)
for contain in pl.containers :
    pl.bar_label(contain)
pl.set_title("Distribution Of The Clusters")
plt.show()
```

## Let's see the multiple clustering model and compare them 

```python
from sklearn.cluster import KMeans,DBSCAN
from sklearn.metrics import silhouette_score
```

```python
# K-Means
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans_labels = kmeans.fit_predict(PCA_data)

# Agglomerative Clustering
agglo = AgglomerativeClustering(n_clusters=4)
agglo_labels = agglo.fit_predict(PCA_data)

# DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(PCA_data)
```

```python
kmeans_silhouette = silhouette_score(PCA_data, kmeans_labels)
agglo_silhouette = silhouette_score(PCA_data, agglo_labels)

# For DBSCAN, exclude noise points labeled as -1
dbscan_silhouette = silhouette_score(PCA_data[dbscan_labels != -1], dbscan_labels[dbscan_labels != -1])

print(f'K-Means Silhouette Score: {kmeans_silhouette}')
print(f'Agglomerative Clustering Silhouette Score: {agglo_silhouette}')
print(f'DBSCAN Silhouette Score: {dbscan_silhouette}')
```

## Based on the silhouette scores, K-Means has the highest silhouette score, indicating it produces the most well-defined clusters compared to Agglomerative Clustering and DBSCAN. Here are the scores for reference:

### K-Means Silhouette Score: 0.4741
### Agglomerative Clustering Silhouette Score: 0.4348
### DBSCAN Silhouette Score: 0.3016

### <font color='Green'>LR (Logistic Regression)</font>

```python
def train_LR_model(x_train, y_train):
    model1 = LogisticRegression(multi_class='ovr').fit(x_train, y_train)
    return model1

def evaluate_model(model1, x_train, y_train, x_test, y_test):
    accuracy = accuracy_score(y_test,model1.predict(x_test))
    LR_precision_macro = precision_score(y_test, model1.predict(x_test), average='macro')
    LR_recall_macro = recall_score(y_test, model1.predict(x_test), average='macro')
    LR_f1_macro = f1_score(y_test, model1.predict(x_test), average='macro')
    return accuracy,LR_precision_macro,LR_recall_macro,LR_f1_macro

def predict_and_compare(model1, x_test, y_test):
    pred = model1.predict(x_test)
    result_pred = pd.DataFrame({"Pred_LR": pred, "Real_Value": y_test.values})
    return result_pred

if __name__ == "__main__":
    model1 = train_LR_model(x_train, y_train)
    accuracy,LR_precision_macro , LR_recall_macro,LR_f1_macro = evaluate_model(model1, x_train, y_train, x_test, y_test)
    print(f"Accuracy Score: {accuracy}")
    print(f"Precision_Macro {LR_precision_macro}")
    print(f"Recall_Macro: {LR_recall_macro}")
    print(f"F1_macro: {LR_f1_macro}")
    LR_pred = predict_and_compare(model1, x_test, y_test)
```
